```python
lista_enlaces=[]
for i in lista_merc:
    lista_enlaces.append(f"https://tienda.mercadona.es/categories/{list(i.values())[0]}")
```

```python
lista_enlaces
```

```python
lista=[]
for x in lista_enlaces:
    url = f"{x}"
    driver.get(url)
    time.sleep(1)
    elements = driver.find_elements_by_class_name("product-cell")
    lista_p=[]
    for i in elements:
        lista_p.append(i.text)
        lista_p.append(x)
    lista.append(lista_p)
    
```

```python
lista2=[]
for i in lista:
    for x in i:
        lista2.append(x)
lista2
```

```python
lista2
sublist = [lista2[n:n+2] for n in range(0, len(lista2), 2)]
sublist
```

```python
mercadona=pd.DataFrame(sublist, columns=["producto","link"])
```

```python
mercadona[["nombre_prod","Q","precio","anadir"]]=mercadona.producto.str.split("\n",expand=True)
mercadona.head()
```

```python
mercadona2= mercadona.replace(to_replace ='[ó]', value = 'o', regex = True) 
mercadona2 = mercadona2.replace(to_replace ='[,]', value = '.', regex = True) 
mercadona2 = mercadona2.replace(to_replace ='[á]', value = 'a', regex = True)
mercadona2 = mercadona2.replace(to_replace ='[é]', value = 'e', regex = True)
mercadona2 = mercadona2.replace(to_replace ='[ó]', value = 'o', regex = True) 
mercadona2= mercadona2.replace(to_replace ='[í]', value = 'i', regex = True)
mercadona2 = mercadona2.replace(to_replace ='[ú]', value = 'u', regex = True) 
mercadona2 = mercadona2.replace(to_replace ='[ñ]', value = 'n', regex = True) 
mercadona2 = mercadona2.replace(to_replace ='aprox', value = '', regex = True) 
```

```python
mercadona2[["precio_medida","medida"]]=mercadona2["precio"].str.split("/",expand=True)
```

```python
mercadona2
```

```python
mercadona2["precio"] = round(pd.to_numeric(mercadona2.precio.replace(to_replace ='[ €]', value = '', regex = True)),3)
```

```python
mercadona2["precio_medida"] = mercadona2.precio_medida.replace(to_replace ='[ €]', value = '', regex = True)
```

```python
mercadona2["medida_rel"]=mercadona2["Q"].str.extract(r"(\d+ [\w.\-]+)",)
mercadona2.head()
```

```python

mercadona2[["q_medida","medida_"]]=mercadona2["medida_rel"].str.split(" ",expand=True)
mercadona2.tail()
```
