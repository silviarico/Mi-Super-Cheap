```python
lista_all=[]
for i in lista_enlaces:
    url = i
    driver.get(url)
    driver.implicitly_wait(2)
    body = driver.execute_script("return document.body")
    source = body.get_attribute('innerHTML') 
    soup_sc = BeautifulSoup(source, "html.parser")
    N=soup_sc.findAll("div",{"class": "pagination-list-and-total"})
    text=[p.getText() for p in N][0]
    num=[int(s) for s in text.split(" ")[1].split() if s.isdigit()][0]
    for x in range(1,num):
        lista_all.append(f"{i}?page={x}&disp=")
    lista_all.append(i)
    
```

```python
lista_productos=[]
for i in lista_all:
    url = i
    driver.get(url)
    driver.implicitly_wait(2)
    body = driver.execute_script("return document.body")
    source = body.get_attribute('innerHTML') 
    soup_sc = BeautifulSoup(source, "html.parser")
    price=soup_sc.findAll("a",{"class": "productMainLink"})
    price_=[p.getText() for p in price]
    for x in price_:
        a=x.replace(',','.')
        a=a.replace("\n\n\n\n\t\t\t\t\t\t\t",",")
        a=a.replace("\xa0€\n",",")
        a=a.replace("\xa0€","")
        a=a.replace(".\n\n\n\n\n","")
        a=a.replace("\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\t\t\t\t\t","")
        a=a.replace("\n\n\n\n\n\n\n\n\n\n\n\t\t\t\t\t","")
        lista=a.split(",")
        lista.append(i)
        lista_productos.append(lista)
```
