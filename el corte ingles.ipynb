```python
url = ('https://www.elcorteingles.es/supermercado/')
driver.get(url)
time.sleep(5)
driver.find_element_by_css_selector('#cookies-agree-all').click()
body = driver.execute_script("return document.body")
source = body.get_attribute('innerHTML') 
soup_sc = BeautifulSoup(source, "html.parser")
driver.quit()
```

```python
paginas_1=soup_sc.findAll("a",{"class": "top_menu-item js-top-menu-item"})
lista_enlaces1=[f"https://www.elcorteingles.es{elemento.get('href')}" for elemento in paginas_1]
```

```python
lista_enlaces1
```

```python
dic={}
for i in lista_enlaces1:
    url = (f"{i}")
    driver = webdriver.Chrome(service=s)
    driver.get(url)
    time.sleep(2)
    driver.find_element_by_css_selector('#cookies-agree-all').click()
    body = driver.execute_script("return document.body")
    source = body.get_attribute('innerHTML') 
    soup_sc = BeautifulSoup(source, "html.parser")
    prod=soup_sc.findAll("li",{"class": "c6-xs"})
    num=[ca.getText() for ca in prod][0].split(" ")
    num=int(num[-1])
    dic[f"{i}"]=num
    driver.quit()
```

```python
lista_productos_alimentacion=[]
for i in lista_alimentacion:
    url = (f"{i}")
    driver = webdriver.Chrome(service=s)
    driver.get(url)
    time.sleep(2.5)
    driver.find_element_by_css_selector('#cookies-agree-all').click()
    time.sleep(1)
    body = driver.execute_script("return document.body")
    source = body.get_attribute('innerHTML') 
    soup_sc = BeautifulSoup(source, "html.parser")
    price=soup_sc.findAll("div",{"class": "product_tile-right_container"})
    price_=[p.getText() for p in price]
    price_.append(i)
    lista_productos_alimentacion.append(price_)
    driver.quit()
```
